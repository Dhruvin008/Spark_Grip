```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
#geeting data of csv in to pandas object
data_link = "http://bit.ly/w-data"
ddata = pd.read_csv(data_link)
ddata.head(5) #printing first 5 data of csv
```

```
#After importing data now we are trying to find relation between Attributes using Graph
ddata.plot(x='Hours', y='Scores', style='*')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

```
print("By seeing graph we easily find that PercentageScore and HoursStudied have relation of +proposnal or we can say That positive corelation.")
```

```
#As we find corelation now we go with the regression method for that first we have to prepare data as model requirment.
X = ddata.iloc[:, :-1].values  
y = ddata.iloc[:, 1].values  
```

```
#now we have to divide data in to two part test and train for that we use predefine method.
from sklearn.model_selection import train_test_split  
from sklearn.linear_model import LinearRegression 
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=0.3, random_state=0) #train,test ratio - 70:30 
```

```
model = LinearRegression()  
model.fit(X_train, y_train) 

print("Model Trained Sucessfull.")
```

```
#Now we will plot the best regression line which is find by the model
rline = model.coef_*X+model.intercept_

# Plotting for the test data
plt.scatter(X, y)
plt.plot(X, rline);
plt.show()
```

```
print(X_test) #testing data 
y_predicted = model.predict(X_test) # Getting predction from regression line
```

```
comparing = pd.DataFrame({'Actual': y_test, 'Predicted': y_predicted})  
comparing 
```

```
#evaluation of model on the basis of mean absolute error
from sklearn import metrics  
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_predicted)) 
```

```
#Precdicting My score based on my Reading hours
my_hours= [[4.5]]
my_score = model.predict(my_hours)
print("Reading Hours = {}".format(my_hours))
print("Predicted Score = {}".format(my_score[0]))
```
